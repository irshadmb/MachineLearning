```python
import tensorflow as tf
```

```python
data = tf.keras.datasets.mnist

(train__images, train_labels), (value_images, value_labels) = data.load_data()
```

```python
import matplotlib.pyplot as plt

plt.imshow(value_images[0], cmap=plt.cm.binary)
plt.show()
```

```python
train__images = train__images / 255.0
value_images = value_images / 255.0

```

```python
model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Flatten(input_shape=(28, 28)),
        tf.keras.layers.Dense(20, activation="relu"),
        tf.keras.layers.Dense(10, activation="softmax"),
    ]
)

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.fit(train__images, train_labels, epochs=5)



```

```python
classified = model.predict(value_images)
print(classified[0])
print(value_labels[0])
```
